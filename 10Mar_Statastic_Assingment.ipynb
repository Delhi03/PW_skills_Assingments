# Estimation Statistics

Estimation statistics involves using sample data to estimate population parameters. It's a fundamental aspect of inferential statistics, where we draw conclusions about a population based on a sample taken from it.

## Point Estimate

- **Definition**: A single value given as an estimate of a population parameter.
- **Example**: The sample mean is a point estimate of the population mean .

## Interval Estimate

- **Definition**: A range of values within which a population parameter is estimated to lie, with a certain level of confidence.
- **Example**: A confidence interval for the population mean might be given as \(\bar{x} \pm \text{Margin of Error}\).

## Illustration

- **Point Estimate**: "The average height of the sample is 170 cm."
- **Interval Estimate**: "We are 95% confident that the average height of the population is between 168 cm and 172 cm."

Estimation statistics helps quantify the uncertainty associated with sample data and provides a more complete picture of the population being studied.


Q2. Write a Python function to calculate mean, std, error of mean,margin of error

```python
import numpy as np
from scipy import stats  # For sem()

# Example dataset
data = np.array( [5, 7, 8, 6, 7, 9, 5, 8, 6, 7])

# Calculate mean
mean_value = np.mean(data)
print(f"Mean: {mean_value:.2f}")

# Calculate standard deviation (sample)
std_dev_sample = np.std(data, ddof=1)
print(f"Sample Standard Deviation: {std_dev_sample:.2f}")

# Calculate standard error of the mean (SEM)
sem_value = stats.sem(data)
print(f"Standard Error of the Mean (SEM): {sem_value:.2f}")

# Calculate margin of error (assuming 95% confidence level, z-score ~1.96)
confidence_level = 0.95
margin_of_error = sem_value * stats.t.ppf((1 + confidence_level) / 2, len(data) - 1)
print(f"Margin of Error (95% CI): ±{margin_of_error:.2f}")
```

Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Degrees of freedom
df = 10

# Generate x values
x = np.linspace(-5, 5, 1000)

# Compute the t-distribution values
y = t.pdf(x, df)

# Plot the t-distribution
plt.plot(x, y, label=f't-distribution (df={df})', color='blue')
plt.title("Student's t-Distribution with 10 Degrees of Freedom")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be 500 with a standard deviation of 50.
Estimate the population mean revenue with a 95% confidence interval.

```python
import scipy.stats as stats
import math

# Given data
sample_mean = 500
sample_std = 50
n = 50
confidence_level = 0.95

# Degrees of freedom
df = n - 1

# Calculate the t critical value for 95% confidence
t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

# Calculate the standard error of the mean
standard_error = sample_std / math.sqrt(n)

# Calculate the margin of error
margin_of_error = t_critical * standard_error

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print(f"Sample Mean: ${sample_mean}")
print(f"95% Confidence Interval: ${confidence_interval[0]:.2f} to ${confidence_interval[1]:.2f}")
```

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

```python
import math
from scipy import stats

# Given data
sample_mean = 10
population_mean = 8
sample_std = 3
n = 100

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / (sample_std / math.sqrt(n))

# Degrees of freedom
df = n - 1

# Calculate the p-value
p_value = stats.t.sf(abs(t_statistic), df) * 2  # two-tailed test

# Output the results
print(f"t-Statistic: {t_statistic:.2f}")
print(f"p-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0). The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis (H0). The sample mean is not significantly different from the population mean.")
```
