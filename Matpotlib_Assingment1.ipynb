Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is widely used in data analysis and scientific research for its ability to produce high-quality plots and charts.

```python
import matplotlib.pyplot as plt
```

```python
print("Line plot")
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

```python
plt.scatter([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

```python
plt.bar(['A', 'B', 'C', 'D'], [10, 20, 25, 30])
plt.show()
```

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
plt.hist(data, bins=4)
plt.show()
```

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels)
plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.   
import numpy as np   
np.random.seed(3)   
x = 3 + np.random.normal(0, 2, 50)   
y = 3 + np.random.normal(0, 2, len(x))  
Note: Also add title, xlabel, and ylabel to the plot.   

Solution:  
    A scatter plot is a type of data visualization that displays individual data points on a Cartesian plane

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("This is title")
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.  
Use the following data:  

Solution:  
    The subplot() function in Matplotlib is used to create multiple plots within a single figure.

```python
x  = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating the first subplot
plt.subplot(2, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y1)
plt.title('Plot 1')

# Creating the second subplot
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Plot 2')

# Creating the third subplot
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Plot 3')

# Creating the fourth subplot
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Plot 4')

# Adjusting the layout
plt.tight_layout()

# Displaying the plot
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.  

Solution:   
A box plot, also known as a box-and-whisker plot, is a statistical graphic used to display the distribution of a dataset. It provides a visual summary of the data, highlighting its central tendency, variability, and potential outliers.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# Creating the box plot
data = [box1, box2]
plt.boxplot(data, labels=['Box 1', 'Box 2'])
```
